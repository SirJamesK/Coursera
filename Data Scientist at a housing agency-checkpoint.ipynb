```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
# Load the dataset
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df = pd.read_csv(boston_url)
```

```python
# Task 1: Provide a boxplot for the "Median value of owner-occupied homes"
plt.figure(figsize=(10, 6))
sns.boxplot(boston_df['MEDV'])
plt.title("Boxplot for the Median Value of Owner-occupied Homes")
plt.xlabel("Median value of owner-occupied homes in $1000's")
```

```python
# Task: Provide a bar plot for the Charles river variable
plt.figure(figsize=(10, 6))
sns.countplot(x='CHAS', data=boston_df)
plt.title("Bar Plot for the Charles River Variable")
plt.xlabel("Charles River dummy variable (1 if tract bounds river; 0 otherwise)")
plt.ylabel("Count")
```

```python
# Task: Provide a boxplot for the MEDV variable vs the AGE variable.
# Discretize the age variable into three groups
boston_df['AGE_GROUP'] = pd.cut(boston_df['AGE'],
                                 bins=[-1, 35, 70, 100],
                                 labels=['35 years and younger', 'between 35 and 70 years', '70 years and older'])

plt.figure(figsize=(10, 6))
sns.boxplot(x='AGE_GROUP', y='MEDV', data=boston_df)
plt.title("Boxplot for the MEDV Variable vs the AGE Variable")
plt.xlabel("Age Group")
plt.ylabel("Median value of owner-occupied homes in $1000's")
```

```python
# Task: Provide a scatter plot to show the relationship between Nitric oxide concentrations
# and the proportion of non-retail business acres per town.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='INDUS', y='NOX', data=boston_df)
plt.title("Scatter Plot of Nitric Oxide Concentrations vs Proportion of Non-retail Business Acres per Town")
plt.xlabel("Proportion of non-retail business acres per town")
plt.ylabel("Nitric Oxides concentration (parts per 10 million)")
```

```python
# Task: Create a histogram for the pupil to teacher ratio variable
plt.figure(figsize=(10, 6))
sns.histplot(boston_df['PTRATIO'], bins=30, kde=True)
plt.title("Histogram for the Pupil to Teacher Ratio Variable")
plt.xlabel("Pupil-Teacher Ratio by Town")
plt.ylabel("Count")
```

```python
# Set the significance level
alpha = 0.05

# T-test for independent samples for the Charles river variable
t_stat, p_val = stats.ttest_ind(
    boston_df.loc[boston_df['CHAS'] == 1, 'MEDV'],
    boston_df.loc[boston_df['CHAS'] == 0, 'MEDV']
)

# ANOVA for the AGE variable
f_stat, p_val_anova = stats.f_oneway(
    boston_df.loc[boston_df['AGE'] <= 35, 'MEDV'],
    boston_df.loc[(boston_df['AGE'] > 35) & (boston_df['AGE'] < 70), 'MEDV'],
    boston_df.loc[boston_df['AGE'] >= 70, 'MEDV']
)

# Pearson Correlation between NOX and INDUS
pearson_coef, p_val_corr = stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])

# Regression analysis for DIS (weighted distances to five Boston employment centres) on MEDV
slope, intercept, r_value, p_val_reg, std_err = stats.linregress(boston_df['DIS'], boston_df['MEDV'])

# Print the results
print("T-test for Charles river variable:")
print("T-statistic:", t_stat)
print("P-value:", p_val)
print("Conclusion:", "Reject H0" if p_val < alpha else "Fail to reject H0", "\n")

print("ANOVA for AGE variable:")
print("F-statistic:", f_stat)
print("P-value:", p_val_anova)
print("Conclusion:", "Reject H0" if p_val_anova < alpha else "Fail to reject H0", "\n")

print("Pearson Correlation between NOX and INDUS:")
print("Pearson Coefficient:", pearson_coef)
print("P-value:", p_val_corr)
print("Conclusion:", "Reject H0" if p_val_corr < alpha else "Fail to reject H0", "\n")

print("Regression analysis for DIS on MEDV:")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r_value)
print("P-value:", p_val_reg)
print("Conclusion:", "Reject H0" if p_val_reg < alpha else "Fail to reject H0")
```
